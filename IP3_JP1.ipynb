5. Remove all the symbols ($) from the 'Price' column and convert and update it to 64-bit int numbers using astype(‘int64’) Hint. Look for string functions like replace/strip).

```python
# str.replace() is used to get rid of the $ symbol and the commas. 
# regex=True is used to ensure that the passed in pattern in a regular expression
# astype('int64') is used to cast a pandas object to a specified data type. the dtype is int64
car_sales['Price'] = car_sales['Price'].str.replace('[\$,]', '', regex=True).astype('int64')
```

6. Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column look up the cumsum() function.

```python
# using cumsum() to take the price from the previous to add tot he price of the next row. A cumulative sum.
car_sales['Total Sales'] = car_sales['Price'].cumsum()
```

```python
# checking to see if Total Sales was added
car_sales.head()
```

7. Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today). Check out “date_range” function and keep the periods=len(car_sales).

```python
# importing datetime 
from datetime import datetime
```

```python
# using date_range to return a range of equally spaced time points
# start is the left bound and is the current date
# period is the number of periods to generate which refers to the length of the columns
# this creates a series of successive dates starting from today
date_range=pd.date_range(start=datetime.now(), periods=len(car_sales))
```

```python
# creating the Sale Date column and setting it equal to date_range
car_sales['Sale Date'] = date_range
```

```python
# checking to see if Sale Date has been added
car_sales.head(6)
```

16. Find out Pearson correlation between 'Total Sales','Price','Odometer (KM)'. Comment on the correlation

```python
heat_map=car_sales[['Total Sales', 'Price', 'Odometer (KM)']].corr(method='pearson')
cols=['Total Sales', 'Price', 'Odometer (KM)']
sns.heatmap(heat_map, xticklabels=cols, yticklabels=cols, annot=True)
```

In conclusion there is a weak positive correlation between price and total sales. There is also a relatively weak negative correlation between the odometer (KM) and total sales. And there is a relatively strong negative correlation between price and the odometer (KM).
